```python
import numpy as np
from numpy import arange
from pylab import plot, xlabel, ylabel, show

# Constants        
Rn_initial = 20000
Po_initial = 0
Pb_214_initial = 0
Bi_initial = 0
Tl_initial = 0
Pb_207_initial = 0

Rn_half_life = 3.82 * 24 * 60 * 60      
Po_half_life = 31 * 60
Pb_214_half_life = 26.8 * 60
Bi_half_life = 19.9 * 60
Tl_half_life = 1.3 * 60

h = 1.0        
p_Rn = 1 - 2**(-h / Rn_half_life)
p_Po = 1 - 2**(-h / Po_half_life)
p_Pb_214 = 1 - 2**(-h / Pb_214_half_life)
p_Bi = 1 - 2**(-h / Bi_half_life)
p_Tl = 1 - 2**(-h / Tl_half_life)

energy_alpha = 4  
energy_r = 3    
energy_beta = 1
energy_z = 7     

num_runs = 10
total_energies = []
energy_alpha_totals = []
energy_beta_totals = []
energy_r_totals = []
energy_z_totals = []

for run in range(num_runs):
    Rn = Rn_initial
    Po = Po_initial
    Pb_214 = Pb_214_initial
    Bi = Bi_initial
    Tl = Tl_initial
    Pb_207 = Pb_207_initial

    energy_alpha_total = 0
    energy_beta_total = 0
    energy_r_total = 0
    energy_z_total = 0

    tmax = 3000000         
    tpoints = arange(0.0, tmax, h)
    Rn_points, Po_points, Pb_214_points, Bi_points, Tl_points, Pb_207_points = ([] for _ in range(6))

    for t in tpoints:  
        Rn_points.append(Rn) 
        Po_points.append(Po)
        Pb_214_points.append(Pb_214)
        Bi_points.append(Bi)
        Tl_points.append(Tl)
        Pb_207_points.append(Pb_207)

        alpha = beta = r = z = 0

        # Tl decay
        decay = sum(np.random.random() < p_Tl for _ in range(int(Tl)))
        Tl -= decay
        Pb_207 += decay
        beta += decay
        energy_beta_total += decay * energy_beta

        # Bi decay
        decay = sum(np.random.random() < p_Bi for _ in range(int(Bi)))
        Bi -= decay

        for _ in range(decay):
            if np.random.random() < 0.997:  # 99.7% to Pb_207
                Pb_207 += 1
                energy_r_total += energy_r
            else:  # 0.3% to Tl
                Tl += 1
                energy_alpha_total += energy_alpha 

        # Pb_214 decay
        decay = sum(np.random.random() < p_Pb_214 for _ in range(int(Pb_214)))
        Pb_214 -= decay
        Bi += decay
        r += decay
        energy_r_total += decay * energy_r

        # Po decay
        decay = sum(np.random.random() < p_Po for _ in range(int(Po)))
        Po -= decay

        for _ in range(decay):
            if np.random.random() < 0.9998:  # 99.98% to Pb_214
                Pb_214 += 1
                energy_alpha_total += energy_alpha 
            else:  # 0.02% to Bi
                Bi += 1
                energy_beta_total += energy_beta 

        # Rn decay
        decay = sum(np.random.random() < p_Rn for _ in range(int(Rn)))
        Rn -= decay
        Po += decay
        z += decay
        energy_z_total += decay * energy_z

    # Total energy calculation
    total_energy = (energy_alpha_total + energy_beta_total + 
                    energy_r_total + energy_z_total)

    total_energies.append(total_energy)
    energy_alpha_totals.append(energy_alpha_total)
    energy_beta_totals.append(energy_beta_total)
    energy_r_totals.append(energy_r_total)
    energy_z_totals.append(energy_z_total)

# Averages and standard deviation
average_total_energy = np.mean(total_energies)
std_total_energy = np.std(total_energies)

average_alpha_energy = np.mean(energy_alpha_totals)
std_alpha_energy = np.std(energy_alpha_totals)

average_beta_energy = np.mean(energy_beta_totals)
std_beta_energy = np.std(energy_beta_totals)

average_r_energy = np.mean(energy_r_totals)
std_r_energy = np.std(energy_r_totals)

average_z_energy = np.mean(energy_z_totals)
std_z_energy = np.std(energy_z_totals)

# Make the graph
plot(tpoints, Rn_points, c='k') 
plot(tpoints, Po_points, c='b') 
plot(tpoints, Pb_214_points, c='r') 
plot(tpoints, Bi_points, c='g')
plot(tpoints, Tl_points, c='c')
plot(tpoints, Pb_207_points, c='m')
xlabel("Time")
ylabel("Number of atoms")
show()

print("Average Total Energy (MeV):", average_total_energy)
print("Standard Deviation of Total Energy (MeV):", std_total_energy)
print("Average Alpha Energy (MeV):", average_alpha_energy)
print("Standard Deviation of Alpha Energy (MeV):", std_alpha_energy)
print("Average Beta Energy (MeV):", average_beta_energy)
print("Standard Deviation of Beta Energy (MeV):", std_beta_energy)
print("Average R Energy (MeV):", average_r_energy)
print("Standard Deviation of R Energy (MeV):", std_r_energy)
print("Average Z Energy (MeV):", average_z_energy)
print("Standard Deviation of Z Energy (MeV):", std_z_energy)
```
